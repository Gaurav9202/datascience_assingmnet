Q4. Explain the difference between the union() and update() methods for a set. Give an example of
each method.

```python
# the main difference between the union() and update() for a set is that it create and return a new set while leaving
# the orignal set unchanged, whereas update() changes the orignal set in place and does not return anything.

a={1,2,3}
b={3,4,5}

# using update()
a.update(b)

b |= a

#now 'a' and 'b' are same
```

Q6. Can we create a nested dictionary? If so, please give an example by creating a simple one-level
nested dictionary.

```python
#Yes, we can create a nested dictonary. 

students = {
    "Alice": {"Math": 90, "Science": 85, "English": 95},
    "Bob": {"Math": 80, "Science": 75, "English": 88},
    "Charlie": {"Math": 92, "Science": 89, "English": 91}
}
students

# print Alice's math grade
print(students["Alice"]["Science"]) # prints 90
```
