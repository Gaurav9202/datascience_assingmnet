Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

A class is a blueprint or template for creating objects (instances). It defines a set of attributes (data members) and methods (functions) that characterize any object of the class. Think of a class as a blueprint that describes the state (attributes) and behavior (methods) that objects of that class will exhibit.


```python
# Define a class named 'Dog'
class Dog:
    # Constructor method to initialize attributes
    def __init__(self, name, age):
        self.name = name    # Attribute: name of the dog
        self.age = age      # Attribute: age of the dog
    
    # Method to make the dog bark
    def bark(self):
        print(f"{self.name} says Woof!")

# Creating an object (instance) of the 'Dog' class
my_dog = Dog("Buddy", 3)

# Accessing attributes and calling methods of the object
print(f"My dog's name is {my_dog.name}.")  # Output: My dog's name is Buddy.
print(f"My dog is {my_dog.age} years old.")  # Output: My dog is 3 years old.
my_dog.bark()  # Output: Buddy says Woof!
```

Q3. Explain why the __init__() function is used. Give a suitable example.

The __init__() function is a special method, also known as the constructor method, which is automatically called when a new instance (object) of a class is created. Its primary purpose is to initialize the attributes (data members) of the object to their initial state.

```python
# Define a class named 'Book'
class Book:
    # Constructor method to initialize attributes
    def __init__(self, title, author):
        self.title = title        # Attribute: title of the book
        self.author = author      # Attribute: author of the book

    # Method to display book details
    def display_details(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")

# Creating an object (instance) of the 'Book' class
book1 = Book("Python Programming", "John Smith")

# Accessing attributes and calling methods of the object
book1.display_details()
```

Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance is a fundamental concept in Object-Oriented Programming (OOP) where a new class (derived class or subclass) is created based on an existing class (base class or superclass). The derived class inherits attributes and methods from the base class, allowing for code reuse and establishing a hierarchical relationship between classes.

Types of inheritance

1.Single Inheritance: It occurs when a subclass inherits from only one superclass.

```python
class Animal:
    def sound(self):
        return "Some sound"

class Dog(Animal):
    def sound(self):
        return "Bark"

my_dog = Dog()
print(my_dog.sound()) 
```

2.Multiple Inheritance:It occurs when a subclass inherits from more than one superclass.

```python
class Bird:
    def fly(self):
        return "Flying"

class Mammal:
    def walk(self):
        return "Walking"

class Bat(Bird, Mammal):
    pass

my_bat = Bat()
print(my_bat.fly()) 
```

3.Multilevel Inheritance:it occurs when a subclass is derived from another subclass.

```python
class Animal:
    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return "Bark"

class Labrador(Dog):
    def speak(self):
        return "Woof"

my_labrador = Labrador()
print(my_labrador.speak()) 
```
