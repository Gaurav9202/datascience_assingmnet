Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator
function.

```python
# A generator function is function that retuns an iterator, which can be iterated over to produce a sequence of value.
# yield keyword is used to reutun stuatement uses for retuning ivalues in python.
```

```python
def generator(n):
    value = 0
    while value < n:
        yield value
        value += 1

for value in generator(5):
    print(value)
```

Q5. Create a generator function for prime numbers less than 1000. Use the next() method to print the
first 20 prime numbers.

```python
def prime_generator():
    for num in range(2, 1000):
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            yield num

prime_gen = prime_generator()
for i in range(20):
    print(next(prime_gen))
```
