14. How to install Python on windows and set path variables.  

To install Python on Windows, download the installer from python.org and run it. During installation, ensure to check "Add Python to PATH" to automatically set up the environment variables. Verify the installation by opening Command Prompt and typing python --version. If Python is not added to PATH, manually add its installation directory to the PATH variable in the system's environment variables settings. This allows Python to be executed from any command prompt window without specifying the full path. Once set up, you can use pip to install Python packages and manage dependencies efficiently.

15. Is indentation required in python ? 

Yes, indentation is required in Python. It uses braces {} to delimit blocks of code, Python uses indentation to define the scope and execution context of statements. Consistent indentation (typically four spaces per level) is crucial for Python's syntax and is enforced by the interpreter.
